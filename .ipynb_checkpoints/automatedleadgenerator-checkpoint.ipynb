# # First

User must run this code below in order to put together a list of names he wants to run through the code. No need
to add anything other than the first and last name. It will ask for the number of doctors first. Then firstname and then lastname over and over until the number of doctors first input is reached.

```python
# creating an empty list 
firstnamelist = [] 
lastnamelist = []
tpsearchlist = []
  
# number of elemetns as input 
n = int(input("Enter number of doctors: ")) 
  
# iterating till the range 
for i in range(0, n): 
    firstname = str(input("Enter first name: "))
    lastname = str(input("Enter last name: "))
    tpsearched = "no"
    tpsearchlist.append(tpsearched)
  
    firstnamelist.append(firstname) # adding the element 
    lastnamelist.append(lastname) # adding the element
      
for i,j,k in zip(firstnamelist,lastnamelist,tpsearchlist):
    print(f"{i} {j} {k}")
```

This is going to be able to take the url from that website with the table and return a list of names so
that the user doesn't have to write out all the names like before. This should be able to make a list of hundreds of names in a moments notice so that the user can use the code below in a moments notice.

```python
firstnamelist = []
lastnamelist = []
doctorphonenumberslistoflists = []

hreffromtplist = []
agefromtplist = []
addressfromtplist = []
emailsfromtplist = []
namefromtplist = []

leadstable = pd.read_csv('indiananames.csv', header=None)
leadstable.columns = ["names", "city", "state", "speaciality", "cert history", "status"]
leadstable
for item in leadstable.names:
    try:
        delimetedname = re.split(' ',str(item))
        last = delimetedname[0]
        first = delimetedname[2]
    except:
        pass
    none = 'none'
    lastnamelist.append(last)
    firstnamelist.append(first)
    doctorphonenumberslistoflists.append(none)
    hreffromtplist.append(none)
    agefromtplist.append(none)
    addressfromtplist.append(none)
    emailsfromtplist.append(none)
    namefromtplist.append(none)
    print(f"{first} {last} {none}")
print(len(firstnamelist))
print(len(lastnamelist))
print(len(emailsfromtplist))
```

## SECOND

In this block of code, the crawler visits the google search results 

```python
#to measure the time of the code
import time

start_time = time.clock()

batchnumber = len(lastnamelist)
doximityurllist = []
healthgradesurllist = []
middleinitiallist = []
statelist = []
certificationlistoflists = []
doctoragelist = []
numberfromdoxlist = []

numberofcertslist = []
statefromhglist = []
numberfromhglist = []
sexfromhglist = []
titlefromhglist = []

secondsrest = 1
#This code uses the doctors name in google searches 
#to grab the first result when searching alongside a certain website we want

#top one is headless bottom one is gui. comment out which one you want.
#browser = webdriver.Chrome(options=options)
browser = webdriver.Chrome()
url = "http://www.google.com"
ticker = 0

for firstname,lastname in zip(firstnamelist,lastnamelist):
    
    #In this portion we are grabbing the results from doximity
    #First we
    browser.get(url)
    searchform = browser.find_element_by_name("q")
    searchform.send_keys(f"Dr {firstname} {lastname} Doximity Neuro")
    searchform.submit()
    #time.sleep(secondsrest) 

    try:
        firstlinkdoximity = browser.find_element_by_xpath(f"//a[contains(.,'doximity') and contains(.,'{firstname}{lastname}')]")
    except:
        try:
            firstlinkdoximity = browser.find_element_by_xpath(f"//a[contains(.,'doximity') or contains(.,'{firstname}') or contains(.,'{lastname}')]")
        except:
            pass
    try:
        hrefdoximity = firstlinkdoximity.get_attribute("href")
        headlinedoximity = firstlinkdoximity.text
        doximityurllist.append(hrefdoximity)
    except:
        doximityurllist.append("none")
        #Next we want to grab the middle initial if we can get it and save that for 
        #True people search later. We will save this href as a bs4 object.
        ####THIS CODE GRABS THE Middle Initial and State
    try:
        page = requests.get(hrefdoximity)
        soup = BeautifulSoup(page.content, 'html.parser')
        middleinitialelement = soup.find("span", {"class":"user-name-middle"})
        middleinitialtext = middleinitialelement.text[0:1]
        #where the list is being appended
        middleinitiallist.append(middleinitialtext)
    except:
        middleinitialtext = "none"
        middleinitiallist.append("none")
    try:   
        stateelement = soup.find("span", {"itemprop":"addressRegion"})
        statetext = stateelement.text
        #where the list is being appended
        statelist.append(statetext)
    except:
        statetext = "none"
        statelist.append("none")
    try:   
        numberfromdoxelement = soup.find("div", {"class":"office-info-telephone"})
        numberfromdoxtext = numberfromdoxelement.text
        numberfromdoxlist.append(numberfromdoxtext)
    except:
        numberfromdoxlist.append("none")
    try:
        certificationsection = soup.find("section", {"class":"section certification-info"})
        certificationlist = certificationsection.find_all("strong")
        #This removes the heading
        certificationlistfixed = certificationlist[1:]
        certificationlisttext = []
        for item in certificationlistfixed:
            certificationlisttext.append(item.text)    
        #where the list is being appended to a greater list
        certificationlistoflists.append(certificationlisttext)
        numberofcerts = len(certificationlisttext)
        numberofcertslist.append(numberofcerts)
    except:        
        certificationlistoflists.append("none")
        numberofcertslist.append("none")
    
    #time.sleep(secondsrest)
    
    #In this portion we are grabbing the results from healthgrades
    
    browser.get(url)
    searchform = browser.find_element_by_name("q")
    if middleinitialtext == "none" and statetext == "none":
        searchform.send_keys(f"Dr {firstname} {lastname} Healthgrades Neuro")
        searchform.submit()   
    elif middleinitialtext == "none":
        searchform.send_keys(f"Dr {firstname} {lastname} Healthgrades {statetext} Neuro")
        searchform.submit() 
    elif statetext == "none":
        searchform.send_keys(f"Dr {firstname} {middleinitialtext} {lastname} Healthgrades Neuro")
        searchform.submit() 
    else:
        searchform.send_keys(f"Dr {firstname} {middleinitialtext} {lastname} Healthgrades {statetext} Neuro")
        searchform.submit()
    time.sleep(secondsrest)
    try:
        firstlinkhealthgrades = browser.find_element_by_xpath(f"//a[contains(.,'healthgrades') and contains(.,'{lastname}')]")
    except:
        pass
    try:
        hrefhealthgrades = firstlinkhealthgrades.get_attribute("href")
        healthgradesurllist.append(hrefhealthgrades)
        
        #hrefdoximity = firstlinkdoximity.get_attribute("href")
        #headlinedoximity = firstlinkdoximity.text
        #doximityurllist.append(hrefdoximity)
    except:
        healthgradesurllist.append("none")
    try:
        ####THIS CODE GRABS THE age and locations
        page = requests.get(hrefhealthgrades)
        soup = BeautifulSoup(page.content, 'html.parser')
        doctorageelement = soup.find("span", {"data-qa-target":"ProviderDisplayAge"})
        doctoragetext = doctorageelement.text[-2:]
        #Where the item is being appended to a list
        doctoragelist.append(doctoragetext)
    except:
        doctoragelist.append("none")
    try:
        page = requests.get(hrefhealthgrades)
        soup = BeautifulSoup(page.content, 'html.parser')
        statefromhgelement = soup.find_all("span", {"itemprop":"title"})[1]
        statefromhgtext = statefromhgelement.text
        statefromhglist.append(statefromhgtext)
    except:
        statefromhglist.append("none")
    try:
        page = requests.get(hrefhealthgrades)
        soup = BeautifulSoup(page.content, 'html.parser')
        titlefromhgelement = soup.find_all("span", {"itemprop":"title"})[0]
        titlefromhgtext = titlefromhgelement.text
        titlefromhglist.append(titlefromhgtext)
    except:
        titlefromhglist.append("none")
    try:
        page = requests.get(hrefhealthgrades)
        soup = BeautifulSoup(page.content, 'html.parser')
        numberfromhgelement = soup.find("a", {"data-hgoname":"phone-number"})
        numberfromhgtext = numberfromhgelement.text
        numberfromhglist.append(numberfromhgtext)
    except:
        try:
            browser.get(hrefhealthgrades)
            browser.find_element_by_xpath('//*[@id="summary-section"]/div/div[1]/div[5]/div[1]/div[2]/div/p[2]/a').click()
            numberfromhgtext = browser.find_element_by_xpath('//*[@id="summary-section"]/div/div[1]/div[5]/div[1]/div[2]/div/p[2]/a').text
            numberfromhglist.append(numberfromhgtext)
        except:
            try:
                browser.get(hrefhealthgrades)
                numberfromhgtext = browser.find_element_by_xpath('//*[@id="summary-section"]/div/div/div[2]/div[1]/div[4]/div[1]/p[2]').text
                numberfromhglist.append(numberfromhgtext) 
            except:
                numberfromhglist.append("none")
    try:
        sexfromhgelement = soup.find("span", {"data-qa-target":"ProviderDisplayGender"})
        sexfromhgtext = sexfromhgelement.text
        sexfromhglist.append(sexfromhgtext)
    except:
        sexfromhglist.append("none")
        
    #time.sleep(secondsrest)
    ticker = ticker + 1
    print(f"{ticker} / {batchnumber} and {hrefhealthgrades}")
    
browser.close()
print("Success!")
print(time.clock() - start_time, "seconds")
```

```python
secondsrest = 2
browser = webdriver.Chrome()
url = "https://www.truepeoplesearch.com/"
browser.get(url)

for i in df.index:
    if df["phone#tp"][i] == "none" and df["middleinitial"][i] == "none" and df["statefromdox"][i] == "none":
#        print(df['lastname'][i])
#        <do scraping code here>
        firstn = df["firstname"][i]
        lastn = df["lastname"][i]
        #state = df["statefromdox"][i]
        try:
            browser.get(url)
            nameform = browser.find_element_by_xpath('//*[@id="Name"]')
            nameform.send_keys(f"{firstn} {lastn}")
            stateform = browser.find_element_by_xpath('//*[@id="CityStateZip"]')
            #stateform.send_keys(f"{state}")
            nameform.submit()
            time.sleep(secondsrest)
    
            try: 
                #this should have an includes text xpath for more accurate selection. Right now it only takes the first result.
                browser.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[4]/div[1]/div[2]/a').click()
            except:
                pass
            try:
                #This clicks for more phones
                browser.find_element_by_xpath('//*[@id="divMorePhones"]/div/button').click()
            except:
                pass
            try:
                hreftruepeoplesearch = browser.current_url
                #This code grabs the phone numbers from truepeoplesearch
                page = requests.get(hreftruepeoplesearch)
                soup = BeautifulSoup(page.content, 'html.parser')
                phoneelements = soup.find_all("a", {"data-link-to-more":"phone"})
                phonelisttext = []
                for item in phoneelements:
                    phonelisttext.append(item.text)    
                #where the list is being appended to an even greater list
                #doctorphonenumberslistoflists.append(phonelisttext)
                df.at[i, 'phone#tp'] = phonelisttext
                #remaining info gets grabbed via xpath
                #tpurl
                try:         
                    df.at[i, 'tpurl'] = hreftruepeoplesearch
                except:
                    df.at[i, 'tpurl'] = 'none'
                #name
                try:
                    namefromtptext = browser.find_element_by_xpath('//*[@id="personDetails"]/div[1]/div/span[1]').text
                    
                    df.at[i, 'namefromtp'] = namefromtptext
                except:
                    df.at[i, 'namefromtp'] = 'none'
                #age
                try:
                    agefromtptext =  browser.find_element_by_xpath('//*[@id="personDetails"]/div[1]/div/span[2]').text
                    
                    df.at[i, 'agefromtp'] = agefromtptext
                except:
                    df.at[i, 'agefromtp'] = 'none'
                #address
                try:
                    currentaddressfromtp = browser.find_element_by_xpath('//*[@id="personDetails"]/div[4]/div[2]/div[2]/div[1]/div/a').text
                    
                    df.at[i, 'addressfromtp'] = currentaddressfromtp
                except:
                    df.at[i, 'addressfromtp'] = 'none'
                #emails
                try:
                    emailelements = soup.find_all("div", {"class":"col-12 col-sm-11"})[4]
                    listofemails = []
                    for item in emailelements:
                        try:
                            emaillink = item.find("div", {"class":"content-value"}).text
                            emaillink = emaillink.rstrip()
                            emaillink = emaillink.lstrip()
                            listofemails.append(emaillink)
                        except:
                            pass

                    df.at[i, 'emailsfromtp'] = listofemails
                except:
                    df.at[i, 'emailsfromtp'] = 'none'
                #end
            except:
                browser.find_element_by_xpath('//*[@id="Name"]').clear()
                browser.find_element_by_xpath('//*[@id="CityStateZip"]').clear()

            browser.find_element_by_xpath('//*[@id="Name"]').clear()
            browser.find_element_by_xpath('//*[@id="CityStateZip"]').clear()
        except:
            pass
        time.sleep(secondsrest)
    elif df["phone#tp"][i] == "none" and df["middleinitial"][i] == "none":
        firstn = df["firstname"][i]
        lastn = df["lastname"][i]
        state = df["statefromdox"][i]
        try:
            browser.get(url)
            nameform = browser.find_element_by_xpath('//*[@id="Name"]')
            nameform.send_keys(f"{firstn} {lastn}")
            stateform = browser.find_element_by_xpath('//*[@id="CityStateZip"]')
            stateform.send_keys(f"{state}")
            nameform.submit()
            time.sleep(secondsrest)
    
            try: 
                #this should have an includes text xpath for more accurate selection. Right now it only takes the first result.
                browser.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[4]/div[1]/div[2]/a').click()
            except:
                pass
            try:
                #This clicks for more phones
                browser.find_element_by_xpath('//*[@id="divMorePhones"]/div/button').click()
            except:
                pass
            try:
                hreftruepeoplesearch = browser.current_url
                #This code grabs the phone numbers from truepeoplesearch
                page = requests.get(hreftruepeoplesearch)
                soup = BeautifulSoup(page.content, 'html.parser')
                phoneelements = soup.find_all("a", {"data-link-to-more":"phone"})
                phonelisttext = []
                for item in phoneelements:
                    phonelisttext.append(item.text)    
                #where the list is being appended to an even greater list
                #doctorphonenumberslistoflists.append(phonelisttext)
                df.at[i, 'phone#tp'] = phonelisttext
                #remaining info gets grabbed via xpath
                #tpurl
                try:         
                    df.at[i, 'tpurl'] = hreftruepeoplesearch
                except:
                    df.at[i, 'tpurl'] = 'none'
                #name
                try:
                    namefromtptext = browser.find_element_by_xpath('//*[@id="personDetails"]/div[1]/div/span[1]').text
                    
                    df.at[i, 'namefromtp'] = namefromtptext
                except:
                    df.at[i, 'namefromtp'] = 'none'
                #age
                try:
                    agefromtptext =  browser.find_element_by_xpath('//*[@id="personDetails"]/div[1]/div/span[2]').text
                    
                    df.at[i, 'agefromtp'] = agefromtptext
                except:
                    df.at[i, 'agefromtp'] = 'none'
                #address
                try:
                    currentaddressfromtp = browser.find_element_by_xpath('//*[@id="personDetails"]/div[4]/div[2]/div[2]/div[1]/div/a').text
                    
                    df.at[i, 'addressfromtp'] = currentaddressfromtp
                except:
                    df.at[i, 'addressfromtp'] = 'none'
                #emails
                try:
                    emailelements = soup.find_all("div", {"class":"col-12 col-sm-11"})[4]
                    listofemails = []
                    for item in emailelements:
                        try:
                            emaillink = item.find("div", {"class":"content-value"}).text
                            emaillink = emaillink.rstrip()
                            emaillink = emaillink.lstrip()
                            listofemails.append(emaillink)
                        except:
                            pass

                    df.at[i, 'emailsfromtp'] = listofemails
                except:
                    df.at[i, 'emailsfromtp'] = 'none'
                #end
            except:
                browser.find_element_by_xpath('//*[@id="Name"]').clear()
                browser.find_element_by_xpath('//*[@id="CityStateZip"]').clear()

            browser.find_element_by_xpath('//*[@id="Name"]').clear()
            browser.find_element_by_xpath('//*[@id="CityStateZip"]').clear()
        except:
            pass
        time.sleep(secondsrest)
    elif df["phone#tp"][i] == "none" and df["statefromdox"][i] == "none":
        firstn = df["firstname"][i]
        middlen = df["middleinitial"][i]
        lastn = df["lastname"][i]
        #state = df["statefromdox"][i]
        try:
            browser.get(url)
            nameform = browser.find_element_by_xpath('//*[@id="Name"]')
            nameform.send_keys(f"{firstn} {middlen} {lastn}")
            stateform = browser.find_element_by_xpath('//*[@id="CityStateZip"]')
            #stateform.send_keys(f"{state}")
            nameform.submit()
            time.sleep(secondsrest)
    
            try: 
                #this should have an includes text xpath for more accurate selection. Right now it only takes the first result.
                browser.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[4]/div[1]/div[2]/a').click()
            except:
                pass
            try:
                #This clicks for more phones
                browser.find_element_by_xpath('//*[@id="divMorePhones"]/div/button').click()
            except:
                pass
            try:
                hreftruepeoplesearch = browser.current_url
                #This code grabs the phone numbers from truepeoplesearch
                page = requests.get(hreftruepeoplesearch)
                soup = BeautifulSoup(page.content, 'html.parser')
                phoneelements = soup.find_all("a", {"data-link-to-more":"phone"})
                phonelisttext = []
                for item in phoneelements:
                    phonelisttext.append(item.text)    
                #where the list is being appended to an even greater list
                #doctorphonenumberslistoflists.append(phonelisttext)
                df.at[i, 'phone#tp'] = phonelisttext
                #remaining info gets grabbed via xpath
                #tpurl
                try:         
                    df.at[i, 'tpurl'] = hreftruepeoplesearch
                except:
                    df.at[i, 'tpurl'] = 'none'
                #name
                try:
                    namefromtptext = browser.find_element_by_xpath('//*[@id="personDetails"]/div[1]/div/span[1]').text
                    
                    df.at[i, 'namefromtp'] = namefromtptext
                except:
                    df.at[i, 'namefromtp'] = 'none'
                #age
                try:
                    agefromtptext =  browser.find_element_by_xpath('//*[@id="personDetails"]/div[1]/div/span[2]').text
                    
                    df.at[i, 'agefromtp'] = agefromtptext
                except:
                    df.at[i, 'agefromtp'] = 'none'
                #address
                try:
                    currentaddressfromtp = browser.find_element_by_xpath('//*[@id="personDetails"]/div[4]/div[2]/div[2]/div[1]/div/a').text
                    
                    df.at[i, 'addressfromtp'] = currentaddressfromtp
                except:
                    df.at[i, 'addressfromtp'] = 'none'
                #emails
                try:
                    emailelements = soup.find_all("div", {"class":"col-12 col-sm-11"})[4]
                    listofemails = []
                    for item in emailelements:
                        try:
                            emaillink = item.find("div", {"class":"content-value"}).text
                            emaillink = emaillink.rstrip()
                            emaillink = emaillink.lstrip()
                            listofemails.append(emaillink)
                        except:
                            pass

                    df.at[i, 'emailsfromtp'] = listofemails
                except:
                    df.at[i, 'emailsfromtp'] = 'none'
                #end
            except:
                browser.find_element_by_xpath('//*[@id="Name"]').clear()
                browser.find_element_by_xpath('//*[@id="CityStateZip"]').clear()

            browser.find_element_by_xpath('//*[@id="Name"]').clear()
            browser.find_element_by_xpath('//*[@id="CityStateZip"]').clear()
        except:
            pass
        time.sleep(secondsrest)
    elif df["phone#tp"][i] == "none":
        firstn = df["firstname"][i]
        middlen = df["middleinitial"][i]
        lastn = df["lastname"][i]
        state = df["statefromdox"][i]
        try:
            browser.get(url)
            nameform = browser.find_element_by_xpath('//*[@id="Name"]')
            nameform.send_keys(f"{firstn} {middlen} {lastn}")
            stateform = browser.find_element_by_xpath('//*[@id="CityStateZip"]')
            stateform.send_keys(f"{state}")
            nameform.submit()
            time.sleep(secondsrest)
    
            try: 
                #this should have an includes text xpath for more accurate selection. Right now it only takes the first result.
                browser.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[4]/div[1]/div[2]/a').click()
            except:
                pass
            try:
                #This clicks for more phones
                browser.find_element_by_xpath('//*[@id="divMorePhones"]/div/button').click()
            except:
                pass
            try:
                hreftruepeoplesearch = browser.current_url
                #This code grabs the phone numbers from truepeoplesearch
                page = requests.get(hreftruepeoplesearch)
                soup = BeautifulSoup(page.content, 'html.parser')
                phoneelements = soup.find_all("a", {"data-link-to-more":"phone"})
                phonelisttext = []
                for item in phoneelements:
                    phonelisttext.append(item.text)    
                #where the list is being appended to an even greater list
                #doctorphonenumberslistoflists.append(phonelisttext)
                df.at[i, 'phone#tp'] = phonelisttext
                #remaining info gets grabbed via xpath
                #tpurl
                try:         
                    df.at[i, 'tpurl'] = hreftruepeoplesearch
                except:
                    df.at[i, 'tpurl'] = 'none'
                #name
                try:
                    namefromtptext = browser.find_element_by_xpath('//*[@id="personDetails"]/div[1]/div/span[1]').text
                    
                    df.at[i, 'namefromtp'] = namefromtptext
                except:
                    df.at[i, 'namefromtp'] = 'none'
                #age
                try:
                    agefromtptext =  browser.find_element_by_xpath('//*[@id="personDetails"]/div[1]/div/span[2]').text
                    
                    df.at[i, 'agefromtp'] = agefromtptext
                except:
                    df.at[i, 'agefromtp'] = 'none'
                #address
                try:
                    currentaddressfromtp = browser.find_element_by_xpath('//*[@id="personDetails"]/div[4]/div[2]/div[2]/div[1]/div/a').text
                    
                    df.at[i, 'addressfromtp'] = currentaddressfromtp
                except:
                    df.at[i, 'addressfromtp'] = 'none'
                #emails
                try:
                    emailelements = soup.find_all("div", {"class":"col-12 col-sm-11"})[4]
                    listofemails = []
                    for item in emailelements:
                        try:
                            emaillink = item.find("div", {"class":"content-value"}).text
                            emaillink = emaillink.rstrip()
                            emaillink = emaillink.lstrip()
                            listofemails.append(emaillink)
                        except:
                            pass

                    df.at[i, 'emailsfromtp'] = listofemails
                except:
                    df.at[i, 'emailsfromtp'] = 'none'
                #end
            except:
                browser.find_element_by_xpath('//*[@id="Name"]').clear()
                browser.find_element_by_xpath('//*[@id="CityStateZip"]').clear()

            browser.find_element_by_xpath('//*[@id="Name"]').clear()
            browser.find_element_by_xpath('//*[@id="CityStateZip"]').clear()
        except:
            pass
        time.sleep(secondsrest)
    else:
        pass

    
browser.close()
```
